```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("./diabetes.csv")
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.columns
```

```python
X=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train, y_train)
```

```python
lr_prediction=lr.predict(X_test)
lr_prediction
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
print(confusion_matrix(y_test, lr_prediction))
print(accuracy_score(y_test, lr_prediction))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5, metric="minkowski", p=2)
knn.fit(X_train, y_train)
```

```python
knn_prediction=knn.predict(X_test)
knn_prediction
```

```python
print(confusion_matrix(y_test, knn_prediction))
print(accuracy_score(y_test, knn_prediction))
```

```python
from sklearn.svm import SVC
svc=SVC(kernel="linear", random_state=0)
svc.fit(X_train, y_train)
```

```python
svc_prediction=svc.predict(X_test)
svc_prediction
```

```python
print(confusion_matrix(y_test, svc_prediction))
print(accuracy_score(y_test, svc_prediction))
```

```python
from sklearn.svm import SVC
kernel_svm=SVC(kernel="rbf", random_state=0)
kernel_svm.fit(X_train, y_train)
```

```python
kernel_svm_prediction=kernel_svm.predict(X_test)
kernel_svm_prediction
```

```python
print(confusion_matrix(y_test, kernel_svm_prediction))
print(accuracy_score(y_test, kernel_svm_prediction))
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(X_train, y_train)
```

```python
nb_prediction=nb.predict(X_test)
nb_prediction
```

```python
print(confusion_matrix(y_test, nb_prediction))
print(accuracy_score(y_test, nb_prediction))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion="entropy")
dt.fit(X_train, y_train)
```

```python
dt_prediction=dt.predict(X_test)
dt_prediction
```

```python
print(confusion_matrix(y_test, dt_prediction))
print(accuracy_score(y_test, dt_prediction))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=20, criterion="entropy")
rf.fit(X_train,y_train)
```

```python
rf_prediction=rf.predict(X_test)
rf_prediction
```

```python
print(confusion_matrix(y_test, rf_prediction))
print(accuracy_score(y_test, rf_prediction))
```
