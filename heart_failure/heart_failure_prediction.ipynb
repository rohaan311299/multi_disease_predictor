```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
dataset=pd.read_csv("./heart_failure_clinical_records_dataset.csv")
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset.columns
```

```python
features = ['age', 'anaemia', 'creatinine_phosphokinase','ejection_fraction', 'high_blood_pressure', 'platelets','serum_creatinine', 'serum_sodium', 'smoking', 'time']
X=dataset[features]
y=dataset['DEATH_EVENT']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.svm import SVC
svc=SVC(kernel="linear", random_state=0)
svc.fit(X_train, y_train)
```

```python
svc_prediction=svc.predict(X_test)
```

```python
svc_prediction
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
print(confusion_matrix(y_test, svc_prediction))
print(accuracy_score(y_test, svc_prediction))
```

```python
from sklearn.linear_model import LogisticRegression
logistic_regression=LogisticRegression(random_state=0)
logistic_regression.fit(X_train, y_train)
```

```python
logistic_regression_prediction=logistic_regression.predict(X_test)
logistic_regression_prediction
```

```python
print(confusion_matrix(y_test, logistic_regression_prediction))
print(accuracy_score(y_test, logistic_regression_prediction))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5, metric="minkowski", p=2)
knn.fit(X_train, y_train)
```

```python
knn_prediction=knn.predict(X_test)
knn_prediction
```

```python
print(confusion_matrix(y_test, knn_prediction))
print(accuracy_score(y_test, knn_prediction))
```

```python
from sklearn.svm import SVC
kernel_svm=SVC(kernel="rbf", random_state=0)
kernel_svm.fit(X_train, y_train)
```

```python
kernel_svm_prediction=kernel_svm.predict(X_test)
kernel_svm_prediction
```

```python
print(confusion_matrix(y_test, kernel_svm_prediction))
print(accuracy_score(y_test, kernel_svm_prediction))
```

```python
from sklearn.naive_bayes import GaussianNB
naive_bayes=GaussianNB()
naive_bayes.fit(X_train, y_train)
```

```python
naive_bayes_prediction=naive_bayes.predict(X_test)
naive_bayes_prediction
```

```python
print(confusion_matrix(y_test, naive_bayes_prediction))
print(accuracy_score(y_test, naive_bayes_prediction))
```

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree=DecisionTreeClassifier(criterion="entropy")
decision_tree.fit(X_train, y_train)
```

```python
decision_tree_prediction=decision_tree.predict(X_test)
decision_tree_prediction
```

```python
print(confusion_matrix(y_test, decision_tree_prediction))
print(accuracy_score(y_test, decision_tree_prediction))
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest=RandomForestClassifier(n_estimators=10, criterion="entropy")
random_forest.fit(X_train,y_train)
```

```python
random_forest_prediction=random_forest.predict(X_test)
random_forest_prediction
```

```python
print(confusion_matrix(y_test, random_forest_prediction))
print(accuracy_score(y_test, random_forest_prediction))
```
